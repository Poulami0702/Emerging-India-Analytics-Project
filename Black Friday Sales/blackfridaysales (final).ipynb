```python
X=df1.drop(['Purchase'], axis=True)
y=df1['Purchase']
```

```python
X.head(1)
```

```python
from sklearn.preprocessing import MinMaxScaler
df_num_scl=MinMaxScaler()
x=df_num_scl.fit_transform(X)
```

```python
x[:1]
```

```python
x=pd.DataFrame(data=x, columns=X.columns)
```

```python
x.head(1)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model=lr.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
import xgboost as xg
from sklearn.metrics import mean_squared_error as MSE
xgb_r = xg.XGBRegressor()
xgb_r.fit(X_train, y_train)
```

```python
xgb_r.score(X_test,y_test)
```

```python
xgb=cross_val_score(xg.XGBRegressor(),X_train, y_train, cv=5)
print(xgb)

print('The mean Score', np.mean(xgb))
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
y_pred = xgb_r.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_pred, y_test))
print("RMSE : ", rmse)
print("R2 score:", r2_score(y_test,y_pred))
```
